# Exploratory Data Analysis (EDA)

```python
def visualize(df, x, y, z, title, type):
  plt.figure(figsize=(20, 5))
  if (type == 'line'):
    plt.plot(x, y, data=df, marker='o')
  elif (type == 'bar'):
    df = df.sort_values(y, ascending=False).head(min(df[x].nunique(), 10))
    plt.bar(x, y, data=df.sort_values(y, ascending=False))
  elif (type == 'grouped-bar'):
    df = df.sort_values(y, ascending=False)
    sns.barplot(x=x, y=y, hue=z, data=df)
  else:
    df = df.sort_values(y, ascending=False).head(min(df[x].nunique(), 10))
    plt.pie(df[y], labels=df[x], autopct='%.2lf%%')
  plt.title(title)
  plt.xlabel(x)
  plt.ylabel(y)
  plt.show()
```

> There are many aspects that can be explored. However, only a few will be selected.

## Daily Trend

```python
eda_trend = df.groupby('Release', as_index=False).agg({'Sales' : 'sum'})
eda_trend.head()
```

```python
visualize(eda_trend, 'Release', 'Sales', '', 'Sales Trend', 'line')
```

> Based on this information, it can be concluded that the highest sales were achieved in 2017, while the lowest sales occurred in several other years. This suggests that specific factors in 2017 might have occurred, such as a major game release, special in-game events, or promotional campaigns, which likely drove increased player spending during that year. Daily trend analysis further reveals patterns like peaks during holidays when players have more free time or surges linked to limited-time offers. Understanding these patterns helps identify what motivates players to spend more, whether it's anticipation for a new title, impulse purchases during sales, or engagement with exclusive content. These patterns also provide valuable understanding of player behavior and spending triggers, allowing differentiation between spending on long-awaited titles, impulse purchases during sales, or engagement with limited-time content.

## Sales

### Sales by Series

```python
eda_series = df.groupby('Series', as_index=False).agg({'Sales' : 'sum'})
eda_series.head()
```

```python
visualize(eda_series, 'Series', 'Sales', '', 'Sales by Series', 'bar')
```

> Based on this information, it can be concluded that the game series with the highest sales are games without a series name, which might be because the games genuinely do not belong to any series due to being standalone titles, very recent releases that haven't spawned sequels yet, or representing diverse smaller titles that haven't been categorized under a unifying series name in the dataset, or simply because the dataset is incomplete. The second highest is the game series Minecraft, likely due to its immense popularity across various platforms, consistent updates and expansions that maintain player engagement, and its appeal to a wide age range, resulting in sustained high sales figures over a long period.

### Sales by Genre

```python
eda_genre = df.groupby('Genre', as_index=False).agg({'Sales' : 'sum'})
eda_genre.head()
```

```python
visualize(eda_genre, 'Genre', 'Sales', '', 'Sales by Genre', 'bar')
```

> Based on this information, it can be concluded that the game genres with the highest sales are games with action role-playing, which might be because this genre often combines engaging combat and character progression with deep storylines and exploration, appealing to a broad audience and encouraging long-term player investment, thus driving higher sales. The second highest is the real-time strategy which actually isn't much different from the first likely because both genres often attract players who enjoy strategic depth, complex decision-making, and a sense of progression or mastery, even though the core gameplay mechanics differ.

### Sales by Developer

```python
eda_developer = df.groupby('Developer', as_index=False).agg({'Sales' : 'sum'})
eda_developer.head()
```

```python
visualize(eda_developer, 'Developer', 'Sales', '', 'Sales by Developer', 'bar')
```

> Based on this information, it can be concluded that the games with the highest sales are games developed by Blizzard Entertainment likely due to their reputation for polished games with engaging gameplay, deep lore, and strong community support. Some popular examples include World of Warcraft, the Diablo series, the StarCraft series, and Overwatch.

### Sales by Publisher

```python
eda_publisher = df.groupby('Publisher', as_index=False).agg({'Sales' : 'sum'})
eda_publisher.head()
```

```python
visualize(eda_publisher, 'Publisher', 'Sales', '', 'Sales by Publisher', 'bar')
```

> Based on this information, it can be concluded that the games with the highest sales are games published by Blizzard Entertainment.
