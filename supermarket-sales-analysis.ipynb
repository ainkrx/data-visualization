## Review Categorical Data Entries

> To shorten the codes, a function will be made so that it can be reused later by only calling the function and insert the parameters needed.

```python
def check_mistypedVal(col_names):
  for i, j in enumerate(col_names):
    print('==============================')
    print((i + 1), j)
    print(df[j].value_counts())
```

```python
col_names = ['Payment', 'Branch', 'City', 'Customer type', 'Gender', 'Product line']
check_mistypedVal(col_names)
```

> Based on this information, it can be concluded that this dataset is well distributed (balance) and one of the categorical variables, 'Gender', has several mistyped values.

```python
dict_gender = {
    'male' : 'Male',
    'female' : 'Female',
    'fml' : 'Female'
}
df = df.replace(dict_gender)
```

```python
check_mistypedVal(col_names)
```

> Hence, it can be concluded that there is no more mistyped values in those categorical variables.

## Remove Duplicates

Hence, to know and remove the duplicates in this dataset, it is necessary to check if there is duplicate in 'Invoice ID' column.

```python
df['Invoice ID'].duplicated().sum()
```

# Exploratory Data Analysis (EDA)

```python
def visualize(df, x, y, z, title, type):
  plt.figure(figsize=(20, 5))
  if (type == 'line'):
    plt.plot(x, y, data=df, marker='o')
  elif (type == 'bar'):
    df = df.sort_values(y, ascending=False).head(min(df[x].nunique(), 10))
    plt.bar(x, y, data=df.sort_values(y, ascending=False))
  elif (type == 'grouped-bar'):
    df = df.sort_values(y, ascending=False)
    sns.barplot(x=x, y=y, hue=z, data=df)
  else:
    df = df.sort_values(y, ascending=False).head(min(df[x].nunique(), 10))
    plt.pie(df[y], labels=df[x], autopct='%.2lf%%')
  plt.title(title)
  plt.xlabel(x)
  plt.ylabel(y)
  plt.show()
```

> There are many aspects that can be explored. However, only a few will be selected.

## Daily Trend

```python
eda_trend = df.groupby('Date', as_index=False).agg({'Quantity' : 'sum', 'Revenue' : 'sum'})
eda_trend.head()
```

```python
visualize(eda_trend, 'Date', 'Revenue', '', 'Revenue Trend', 'line')
```

> Based on this information, it can be concluded that the highest Revenue is achieved on the first week of February and the lowest Revenue is achieved on the second week of February. This may be due to several factors, as example, promotional campaigns or special events occurring during the first week that boosted sales, while the second week might have experienced a natural decline in customer activity following the initial surge. Additionally, external factors like holidays, market competition, or inventory availability could have influenced these fluctuations in revenue.

## Payment Methods

```python
eda_payment = df.groupby('Payment', as_index=False).agg({'Revenue' : 'sum', 'Invoice ID' : 'nunique'})
eda_payment.head()
```

```python
visualize(eda_payment, 'Payment', 'Invoice ID', '', 'Transaction Count by Payment', 'pie')
```

> Based on this information, even though there are still a lot of transactions at this store are still made with cash, but it can be concluded that many people now prefer cashless payments. As example, customers usually pay according to the transaction amount. For large amounts, credit cards are commonly used because they offer convenience and security for bigger purchases. For smaller amounts, cash or e-wallet payments are preferred since they are quicker and simpler. When promotions are available, customers tend to favor e-wallets or credit cards because these methods often provide additional discounts or cashback.

## Branches

```python
eda_branch = df.groupby(['Branch', 'Payment'], as_index=False).agg({'Invoice ID' : 'nunique'})
eda_branch.head()
```

```python
visualize(eda_branch, 'Branch', 'Invoice ID', 'Payment', 'Branch by Payment', 'grouped-bar')
```

> Based on this information, it can be concluded that the distribution of payment method usage across branches appears fairly balanced, with no significant disparities. Since branches are not always located in different cities and multiple branches can exist within the same city, this balanced distribution suggests consistent payment behavior within those areas. Payment behavior can be influenced by factors such as local familiarity with e-wallets, infrastructure availability, and demographic differences. For example, in some areas, people may rarely use e-wallets because they are not yet familiar with the technology, leading to uneven adoption rates. Therefore, the observed balanced distribution likely indicates that these branches have achieved a reasonable level of payment method adoption across their respective locations.
